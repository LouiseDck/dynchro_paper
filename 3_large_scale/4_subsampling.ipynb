# 1. Subsampling functions

```python
def sample_uniform(dataset, amount):
    indices = np.random.choice(dataset.obs.index, amount, replace=False)
    return dataset[indices]
```

```python
def divide_dataset(dataset, nr):
    # do not use the same cells multiple times, remove them
    allindices = dataset.obs.index
    size = len(dataset.obs.index) // nr

    divisions_indices = []
    datasets = []

    for i in range(nr):
        indices = np.random.choice(allindices, size, replace=False)
        divisions_indices.append(indices)
        allindices = np.setdiff1d(allindices, indices)
        datasets.append(dataset[indices])

    return datasets, divisions_indices
```
