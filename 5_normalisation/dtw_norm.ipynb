```python
import numpy as np
import dynchro

# get random matrix of 100x100
# get second random matrix of 100x100

nr_cells = [i * 100 for i in range(1, 20)]
results = []

for i in nr_cells:
    mtx1 = np.random.rand(i, 500)
    mtx2 = np.random.rand(i*2, 500)

    res = dynchro.tl.dtw(mtx1, mtx2, mode = "only_results")
    results.append(res[0])

    print(f"DTW distance for matrices of size {i}x{i}: {res[0]}")
```

```python
import matplotlib.pyplot as plt
sizes = [i * 100 for i in range(1, 20)]
distances = results
distances_norm = [d / (s+(2*s)) for d, s in zip(distances, sizes)]

plt.plot(sizes, distances)
plt.xlabel('Matrix Size (n x n)')
plt.ylabel('DTW Distance')
plt.title('DTW Distance vs. Matrix Size')
plt.grid(True)
plt.show()

plt.plot(sizes, distances_norm)
plt.xlabel('Matrix Size (n x n)')
plt.ylabel('Normalized DTW Distance')
plt.title('Normalized DTW Distance vs. Matrix Size')
plt.grid(True)
plt.show()
```
